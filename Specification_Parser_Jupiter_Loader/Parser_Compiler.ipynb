# Generating Jupyter notebook out of compiler

```python
nb = nbf.v4.new_notebook()
```

Next code cell generates Jupyter output out of compiler. The string formatting is semi-hard-coded (the parameters are loaded, but the order of filling in the string format is not), I am currently debugging the fully automated generator

```python
text = '## Statement \n' + test.statement.format(sp.latex(test.Function_instance), test.point)

solution = '## Solution \n' + test.solution.format(sp.latex(test.Function_instance), sp.latex(sp.diff(test.Function_instance,'x')) ,test.point,
                                                  test.point, sp.latex(sp.diff(test.Function_instance,'x').subs('x',test.point)))


code = """\
#Your answer goes here
f'= 

sp.diff(test.Function_instance,'x').subs('x',test.point) == f';"""


nb['cells'] = [nbf.v4.new_markdown_cell(text),
               nbf.v4.new_code_cell(code),
              nbf.v4.new_markdown_cell(solution)]
```

```python
nbf.write(nb, 'test.ipynb')
```
